# Criação de um DataFrame

Este material visa a documentação/aprendizagem sobre análise exploratória de um conunto de dados.

A criação de um DataFrame começa na manipulação de um *dicionário* , em que as chaves possuem uma lista de elementos de igual tamanho. Assim, as chaves reprentam as colunas e os valores de sua lista representam o valor da linha.

```python
import pandas as pd
```

O exemplo utilizado são os 6 cursos de cada campus da UFSM, note que não foi escolhido todos os cursos devido a quantidade maior do campus sede (Santa Maria).

```python
import pandas as pd

cursos = pd.DataFrame ({
    "Santa": [ "Artes Cênicas", "Ciências Biológicas", "Administração", "Agronegócio","Agronomia", "Zootecnia"],

    "Cachoeira do": ["Arquitetura e Urbanismo", "Ciência de Dados e Inteligência Artificial","Engenharia Agrícola","Engenharia Elétrica","Engenharia Mecânica", "Engenharia de Transportes e Logística"],

    "Frederico": [ "Agronomia", "Direito", "Engenharia Ambiental e Sanitária", "Engenharia Florestal","Jornalismo", "Sistemas de Informação"],

    "Palmeira das": [ "Administração", "Ciências Biológicas", "Ciências Econômicas", "Enfermagem", "Nutrição", "Zootecnia" ]})

```

```python
cursos
```

## Informações sobre o DataFrame

```python
cursos.info()
```

```python
cursos.columns
```

```python
cursos.describe()
```

## Renomear nomes das tabelas

```python
cursos_renomeado = cursos.rename(columns={
    "Santa" : "Santa Maria",
    "Cachoeira do" : "Cachoeira do Sul",
    "Frederico" : "Frederico Westphalen",
    "Palmeira das" : "Palmeira das Missões" })
```

```python
cursos_renomeado
```

## Acessando uma coluna

```python
cursos_renomeado['Palmeira das Missões']
```

## Selecionando o índice [1] do DataFrame

```python
cursos_renomeado.iloc[1]
```

## Criação de Series

Criação de DataFrame a partir de uma lista de elementos

```python
pd.Series([80, 6, 7, 6, 11])
```

Alteraração dos nomes dos índices e o nome da Series:

```python
pd.Series([80, 6, 7, 11], index = ["Campus Sanata Maria","Campus Cachoeira do Sul","Campus Frederico Westphalen","Campus Palmeira das Missões"], name= "Quantidade de Cursos por Campus da UFSM")
```

# Criando novas colunas

A criação de uma nova coluna é atráves de uma atribuição de Series de valores a uma nova chave no DataFrame.

```python
cursos_renomeado["Campus"] = "UFSM"
```

```python
cursos_renomeado
```

Removendo uma coluna

```python
del cursos_renomeado["Campus"]
```

```python
cursos_renomeado
```

# Seleção Condiconal: Uso do método query

Ela filtra lnhas de um DataFrame baseado em uma query (pergunta).
