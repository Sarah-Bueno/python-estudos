# Mini Projeto  - Simulação de Fluxo de Veículos com NumPy

**O objetivo deste mini projeto é introduzir manipulação de matrizes com a biblioteca NumPy, teremos a análise em um cenário sobre fluxo de veículos.**

#1.Importando a Biblioteca NumPy

```python
import numpy as np
```

#2.Cruzamento entre vias

Cenário: fluxo de veículos observado em três períodos do dia e três vias urbanas.

linha: representa um período de tempo

coluna: representa uma via

```python
fluxo = np.array([
    [120, 95, 80],
    [100, 140, 90],
    [85, 110, 130]
])
```

#3.Informações do fluxo

**Qual o fluxo total?**

```python
fluxo.sum()
```

O fluxo total observado foi de 950 veículos considerando todas as vias e períodos analisados.

**Qual via recebe mais veículos?**

A segunda via apresentou o maior fluxo acumulado.

```python
fluxo.sum(axis=0)
```

A segunda via apresentou o maior fluxo acumulado, indicando maior concentração de veículos durante o período analisado.

**Qual horário teve maior movimento?**

```python
fluxo.sum(axis=1)
```

O segundo período apresentou o maior volume de tráfego, podendo representar um horário de pico.

**Qual a média de veículos por via?**

```python
fluxo.mean(axis=0)
```

A média de veículos permite identificar quais vias recebem maior fluxo ao longo do tempo.

**Qual foi o ponto mais congestionado?**

```python
fluxo.max()
```

#4.Manipulação de dados do fluxo

**E se o trânsito crescer 20%?**

```python
fluxo_futuro = fluxo * 1.20
```

```python
fluxo_futuro
```

Assim, comparamos o fluxo original com o aumento de 20%

```python
fluxo.sum()
fluxo_futuro.sum()
```

**E se houver uma melhoria em uma das vias?**

```python
fluxo_melhoria = fluxo.copy()
```

assim, temos uma redução de 15% do fluxo de uma das vias

```python
fluxo_melhoria[:,1] = fluxo_melhoria[:,1] * 0.85
```

```python
fluxo_melhoria
```

# 5.Visualização

**Importar uma biblioteca de visualização de dados**

```python
import matplotlib.pyplot as plt
```

**Detalhamos os períodos analisados e seu fluxo no gráfico.**

```python
plt.plot(fluxo.sum(axis=1), marker="o")
plt.title("Fluxo de veículos por período")
plt.xlabel("Período")
plt.ylabel("Quantidade de veículos")
plt.grid()
plt.show()
```

O gráfico demonstra a variação do fluxo de veículos entre os períodos analisados, permitindo identificar momentos de maior movimentação e possíveis horários de pico.

#6.Conclusão

A utilização de matrizes NumPy permitiu representar e analisar fluxos de veículos de forma eficiente, foram simulados cenários de crescimento do tráfego e possíveis mudanças viárias, demonstrando como estruturas matriciais podem ser utilizadas em estudos introdutórios de mobilidade urbana.
